```python
import pandas as pd
```

```python
df = pd.read_csv('healthcare_dataset.csv')
df.head()
```

```python
df['Blood Type'].unique()
```

```python
df2 = df.groupby(['Blood Type']).count()
```

```python
df2.head()
```

```python
df2.reset_index(inplace = True)
```

```python
df2.head()
```

```python
df2 = df2[['Blood Type', 'Name']]
df2.head()
```

```python
tamao = len(df2)
tamao
```

```python
datos = '['
datos
```

```python
i = 1
for index, row in df2.iterrows():
  blood = row['Blood Type']
  value = row['Name']
  if i < tamao:
    datos += '{ "blood type": "'+blood+'", "value": '+str(value)+'},'
    i = i + 1
  else:
    datos += '{ "blood type": "'+blood+'", "value": '+str(value)+'} ]'
```

```python
datos
```
