```python
import pandas as pd
```

```python
df = pd.read_csv('mexico_covid19.csv')
df.head()
```

```python
df['ENTIDAD'].unique()
```

```python
len(df['ENTIDAD'].unique())
```

```python
df2 = df.groupby(['ENTIDAD']).count()
```

```python
df2.head()
```

```python
df2.reset_index(inplace = True)
```

```python
df2.head()
```

```python
df2 = df2[['ENTIDAD', 'id']]
df2.head()
```

```python
tamaño = len(df2)
tamaño
```

```python
datos = '['
datos
```

```python
i = 1
for index, row in df2.iterrows():
  entidad = row['ENTIDAD']
  value = row['id']
  if i < tamaño:
    datos += '{ "entidad": "'+entidad+'", "value": '+str(value)+'},'
    i = i + 1
  else:
    datos += '{ "entidad": "'+entidad+'", "value": '+str(value)+'} ]'
```
